```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
# plt.scatter on pandas data

df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.title('Avg and SR analysis of top 50 batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red')
plt.title('Avg and SR analysis of top 50 batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
#marker
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='blue',marker='+')
plt.title('Avg and SR analysis of top 50 batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

# Pie Chart

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data 

data = [23,45,100,20,49]
subjects =['eng','science','maths','sst','hindi']

plt.pie(data)
plt.show()
```

```python
plt.pie(data,labels=subjects)
plt.show()
```

```python
#data set

df4 = pd.read_csv('gayle-175.csv')
df4
```

```python
plt.pie(df4['batsman_runs'])

plt.show()
```

```python
plt.pie(df4['batsman_runs'],labels=df4['batsman'])

plt.show()
```

```python
plt.pie(df4['batsman_runs'],labels=df4['batsman'],autopct='%0.1f%%')

plt.show()
```

```python
# percentage and colors

plt.pie(df4['batsman_runs'],labels=df4['batsman'],autopct='%0.1f%%',colors=['blue','black','yellow','green','pink','brown'])
plt.show()
```

```python
# explode shadow

plt.pie(df4['batsman_runs'],labels=df4['batsman'],autopct='%0.1f%%',colors=['blue','black','yellow','green','pink','brown'],explode=[0.3,0,0,0,0,0.1])
plt.show()
```

```python
plt.pie(df4['batsman_runs'],labels=df4['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
